# Step: Create a Query Engine and Ask Questions


```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.core.settings import Settings

# ✅ Set a compatible embedding model
Settings.embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2")

# ✅ Use correct model and tokenizer, remove 'temperature'
Settings.llm = HuggingFaceLLM(
    model_name="tiiuae/falcon-rw-1b",
    tokenizer_name="tiiuae/falcon-rw-1b",
    context_window=2048,  # Full model capacity
    max_new_tokens=200,   # Leave enough room for PDF + query
    generate_kwargs={"top_p": 1.0},
    device_map="auto"
)


# ✅ Load your documents and build the index
documents = SimpleDirectoryReader("data").load_data()
index = VectorStoreIndex.from_documents(documents)

# ✅ Run a test query
query_engine = index.as_query_engine()
response = query_engine.query("What are the types of skin cancer?")
print(response)
```
